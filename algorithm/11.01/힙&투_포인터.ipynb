# **heap**

* **완전 이진 트리**의 일종으로 **우선순위 큐**를 위해 만들어진 자료구조
* 목적: 여러 값들 중 **최댓값 또는 최솟값**을 빠르게 탐색
* **반정렬 상태(느슨한 정렬)** 유지

### **1. 구조**
> *  부모 노드는 자식 노드보다 항상 큰(작은) 값을 유지
> *  max heap - key(parent) >= key(child)
> *  min heap - key(parent) <= key(child)

### **3. heapq 패키지**

> * 최소 힙 자료구조를 지원하는 파이썬 내장 모듈
> * heappush, heappop, heapify 등 메서드 지원

```python
# 모듈 Import
from heapq import heappush, heappop

# 최소 힙 생성
heap = []

# 원소 초가
heappush(heap, 3)
heappush(heap, 2)
heappush(heap, 1)
print("HEAP: ", heap)

# 원소 제거
print(heappop(heap))
print(heappop(heap))
print("HEAP: ", heap)
```

```python
from heapq import heapify

# 리스트 -> 힙 자료구조
heap = [1, 5, 2, 9, 5, 3, 7, 6]

heapify(heap)

print("HEAP: ", heap)

# 최소값 탐색
print("MINIMUM VALUE: ", heap[0])
```

```python
from random import shuffle
import time

stud_id = ["2022"+str(x) for x in range(10000, 20000)]
ranks = [x+1 for x in range(10000)]
shuffle(ranks)

profiles = [(ranks[x], stud_id[x]) for x in range(len(ranks))]
print(profiles)

# 1~3등 학번 찾기

## 내장 함수 활용
start = time.time()
rankers = []
for i in range(3):
  #ranker = min([x[1] for x in profiles])
  ranker = min(profiles, key=lambda x: x[0])
  profiles.remove(ranker)
  rankers.append(ranker)
print(rankers)
end = time.time()
print(end-start, "sec")
print('===========================')

profiles = [(stud_id[x], ranks[x]) for x in range(len(ranks))]
print(profiles)

## 우선순위큐로 찾기
start = time.time()
heapify(profiles)
rankers=[]
for i in range(3):
  rankers.append(profiles[i])
print(rankers)
end = time.time()
print(end-start, "sec")
```
